## Data description
The business meaning of each column in the data is as below

* <b>iyear</b>: The year of the terrorist attack , <b>ordinal categorical</b>
* <b>imonth</b>: The month of the terrorist attack ,<b>ordinal categorical</b>
* <b>iday</b>: The day of the attack in the respective month,<b>ordinal categorical</b>
* <b>country_txt</b>: Country which was the victim of the attack,<b>nominal categorical</b>
* <b>region_txt</b>: The geographic region in which the country falls,<b>nominal categorical</b>
* <b>provstate</b>: The political division of the country in which the attack took place,<b>nominal categorical</b>
* <b>city</b>: The city where the attack took place,<b>nominal categorical</b>
* <b>multiple</b>: 1 If Multiple no of attacks took place at the same day, 0 otherwise,<b>binary categorical</b>
* <b>success</b>: 1 if the attack planned was successfully operated,else 0,<b>binary categorical</b>
* <b>suicide</b>: 1 if it was a suicide attack, else 0,<b>binary categorical</b>
* <b>attacktype1_txt</b>: The type of attack,<b>nominal categorical</b>
* <b>targtype1_txt</b>: The community targeted for attack,<b>nominal categorical</b>
* <b>targsubtype1_txt</b>: The sub community specifically targeted for attack,<b>nominal categorical</b>
* <b>gname</b>: The terrorist group involved,<b>nominal categorical</b>
* <b>corp1</b>: The individual corporation targeted,<b>nominal categorical</b>
* <b>target1</b>: The people/group targeted for the attack,<b>nominal categorical</b>
* <b>natlty1_txt</b>: The nation to which the attackers belong,<b>nominal categorical</b>
* <b>weaptype1_txt</b>: The weapon type used in the attack,<b>nominal categorical</b>
* <b>weapsubtype1_txt</b>: The sub category of weapon used in the attack,<b>nominal categorical</b>
* <b>nkill</b>: The number of people killed in the attack,<b>ordinal categorical</b>
* <b>nwound</b>: The number of people got wounded in the attack,<b>ordinal categorical</b>



# Missing values treatment
* Impute the missing values with MODE value for categorical variables
* Interpolate the values based on nearby values
* Interpolate the values based on business logic

```python
terrorism_data.isnull().sum()
```

```python
terrorism_data.columns
```

```python
# Treating missing values of categorical variable with MODE value
terrorism_data['provstate'].fillna(value="Not Specified", inplace=True)
terrorism_data['city'].fillna(value="Not Specified", inplace=True)
terrorism_data['multiple'].fillna(value=terrorism_data['multiple'].mode()[0], inplace=True)
terrorism_data['targsubtype1_txt'].fillna(value="Data not found", inplace=True)
terrorism_data['corp1'].fillna(value="Not Specified", inplace=True)
terrorism_data['target1'].fillna(value="Not Specified", inplace=True)
terrorism_data['natlty1_txt'].fillna(value="Not Specified", inplace=True)
terrorism_data['weapsubtype1_txt'].fillna(value="Not Specified", inplace=True)
terrorism_data['nkill'].fillna(value=terrorism_data['nkill'].mode()[0], inplace=True)
terrorism_data['nwound'].fillna(value=terrorism_data['nwound'].mode()[0], inplace=True)
```

```python
def PlotBarCharts(inpData, colsToPlot):
    %matplotlib inline
    
    import matplotlib.pyplot as plt
    
    # Generating multiple subplots
    fig, subPlot=plt.subplots(nrows=1, ncols=len(colsToPlot), figsize=(55,10))
    fig.suptitle('Top Terrorists Incidents by: '+ str(colsToPlot))
    colours=["Blues","cubehelix","YlOrBr"]
    for colName, plotNumber,colorname in zip(colsToPlot, range(len(colsToPlot)),colours):
        if(colName=='city'):
            df1 = inpData[inpData[colName] != 'Unknown']
            Grouped=df1.groupby(colName).size().sort_values(ascending=False).head(10)
            
        else:
            Grouped=inpData.groupby(colName).size().sort_values(ascending=False).head(10)
        
        x_axis=Grouped.index.to_list()
        Y_axis=Grouped
        barplots=sns.barplot(x = x_axis,y = Y_axis,palette = colorname ,ax=subPlot[plotNumber])
        barplots.set_xticklabels(labels=x_axis,rotation=30)
        barplots.set(xlabel=colName, ylabel = ' No of attacks ',title=' Terror strikes by '+ colName)
```

```python
# Calling the function
PlotBarCharts(inpData=terrorism_data, colsToPlot=['country_txt', 'provstate', 'city'])
```
